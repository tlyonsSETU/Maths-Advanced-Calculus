# Example 1
* During Lecture (Thursday, 9 March 2023) we saw that the flux integral for the vector field
$$
\vec{F}(x,y) = x^2y\hat{\imath} + y^2x\hat{\jmath}
$$
along the boundary of the region
$$
R:-1\leq x\leq 1\quad -1\leq y\leq 1
$$
was zero.

* To see why this is the case, we are going to use __plt.quiver()__ to plot the vector field $\vec{F}(x,y)$ and the boundary $\partial R$.


* The __quiver__ function has the following usage: __quiver(x,y,u,v)__
    * __x__ and __y__ are the coordinates of each vector in the vector field.
    * __u__ and __v__ are the horizontal and vertical components respectively,at each of these locations

### Defining the vector field
* In the cell below, we __def__ine functions __F1(x,y)__ and __F2(x,y)__ to represent the components of the vector field $\vec{F}(x,y)$, so we can use them in the quiver function. 

```python
def F1(x,y):
  return x**2*y

def F2(x,y):
  return y**2*x
```

### Setting up the (x,y)-plane

* In the cell below we use __np.linspace(-1,1,25)__ to create a list of 25 numbers, starting as -1 and ending at +1.

* We create two of these lists, one to represent the x-axis and another to represent the y-axis.


* We use the command __np.meshgird(X,Y)__ to combine these lists into a __two-dimensional array__ of numbers we call __x,y__, which represent points in the (x,y)-plane. 

```python
X=np.linspace(-1,1,25)
Y=np.linspace(-1,1,25)
x,y=np.meshgrid(X,Y)
```

## Plotting the vector field

* The cell below creates a __figure__ environment to contain our plot.


* The command __plt.quiver()__ plot the functions __F1(x,y)__ and __F2(x,y)__ over the x,y-meshgrid we created above.


* The command __plt.axhline()__ plots a __h__orizontal line at the specified hieght __y__.


* The command __plt.axvline()__ plots a __v__ertical line at the specified location __x__.

```python
plt.figure(figsize=(12,12))

plt.quiver(x , y , F1(x,y) , F2(x,y),scale=2,scale_units='xy',angles='xy')  # The quiver plot for the gradient field

plt.axhline(y=-1,c='b',lw=2,alpha=0.5) # Plot the lower bdy. of R
plt.axhline(y=1,c='b',lw=2,alpha=0.5) # Plot the upper bdy. of R

plt.axvline(x=-1,c='b',lw=2,alpha=0.5) # Plot the left bdy. of R
plt.axvline(x=1,c='b',lw=2,alpha=0.5) # Plot the right bdy. of R


plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.xlabel('x',size=12) # The x label

plt.ylabel('y',size=12) # The y label
```

## Observations

1. We can see from this plot that sometimes the vector field flows into to region $R$ and sometimes it flows out.


2. We can see that along any bouundary, the vector field flows __in__ through one half of the line and __out__ through the other half.


3. Overall, there is a __net zero__ flow through each boundary, i.e. as much flows in as flows out through __each__ boundary.


4. This is why the __flux integrals__ were __zero__ along each boundary when they calculated in lectures.
